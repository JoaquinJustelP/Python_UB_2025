# **Visualización de datos**

La visualización de datos es una técnica clave en el análisis de datos y la ciencia de datos en general. A menudo, los datos pueden ser complejos y difíciles de entender cuando se presentan en forma de tablas y números. Sin embargo, cuando se representan de manera visual, los patrones y tendencias se hacen más claros y fáciles de interpretar.

En python existen varias librerías para la visualización de datos, algunos ejemplos son:
* Matplotlib
* Seaborn
* Plotly
* ...

## **¿Por qué es importante la visualización de datos?**

La visualización de datos es fundamental en el análisis y la ciencia de datos porque permite:

- **Detectar patrones y tendencias** que no son evidentes en tablas o listas de números.
- **Comunicar resultados** de manera clara y efectiva a diferentes audiencias.
- **Identificar valores atípicos, errores o anomalías** en los datos.
- **Explorar relaciones** entre variables y facilitar la toma de decisiones basada en datos.

Una buena visualización puede transformar datos complejos en información comprensible y útil.



## **Buenas prácticas en visualización de datos**

Al crear visualizaciones, es importante seguir ciertas buenas prácticas para asegurar que los gráficos sean claros y útiles:

- **Elige el tipo de gráfico adecuado** para el tipo de datos y el mensaje que quieres transmitir.
- **Evita la sobrecarga de información**: no incluyas demasiados elementos en un solo gráfico.
- **Usa colores y estilos de forma coherente** y accesible (considera el daltonismo).
- **Incluye títulos, etiquetas y leyendas** para que el gráfico sea autoexplicativo.
- **No distorsiones los datos**: mantén proporciones y escalas adecuadas.
- **Simplifica**: elimina elementos innecesarios que no aporten información relevante.

Una visualización efectiva debe ser precisa, clara y centrada en el mensaje que se quiere comunicar.

#### **Ajustes del plot**

Podemos cambiar varios parámetros de la gráfica para personalizarlo a nuestro gusto, por ejemplo, el estilo o los colores de las líneas.

Para los colores:


```python
plt.plot(x, np.sin(x - 0), color='blue')        # especificar color por nombre
plt.plot(x, np.sin(x - 1), color='g')           # código del color (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # escala de grises de 0 a 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Codigo Hex  (RRGGBB desde 00 a FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, valores de 0 a 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # todos los nombres de colores HTML
```

Lo mismo para las lineas:

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# FO con los siguientes códigos:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

O combinando ambos:

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

#### **Gráfico de dispersión**

Los gráficos de dispersión o "scatter plots" los podemos hacer con la función `plt.scatter`.



```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.scatter(x, y, marker='o');
```

Esta función nos permite ajustar las propiedades de los diferentes puntos, como el color, el tamaño, la forma...

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis') #alpha controla la transparencia
plt.colorbar();  # para mostrar la escala de colores
```

También podemos utilizar estos argumentos para mostrar cierta información de interes, para así poder observar datos multidimensionales.
Por ejemplo, usando el dataset de Iris:

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

Podemos ver que esta gráfica de dispersión nos ha dado la capacidad de explorar simultáneamente cuatro dimensiones diferentes de los datos: la ubicación (x, y) de cada punto corresponde a la longitud y el ancho de la sépala, el tamaño del punto está relacionado con el ancho del pétalo, y el color está relacionado con las especies particulares de flores.

#### **Gráficas de densidad y contorno**

A veces es útil mostrar datos tridimensionales en dos dimensiones utilizando contornos o regiones codificadas por colores. Para ello, Matplotlib cuenta con varias funciones como `plt.contour` o `plt.contourf`.


`plt.contour` toma tres argumentos, un grid de valores x, un grid de valores y y un grid de valores z. Los x e y representan las posiciones en el plot, y los z representan los niveles de contorno.

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y) #para crear el grid de 2d
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black');
```

Las lineas continuas son valores positivos y las discontinuas, negativos. Podemos cambiar la manera de representar esto, por ejemplo codificandolo por colores, añadiendo el argumento `cmap` para especificar el "colormap".

También podemos modificar el numero de lineas a dibujar, en este caso queremos 20 intervalos igualmente espaciados dentro del rango de datos.

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

Podemos rellenar los espacios entre las líneas utilizando la función `plt.contourf`

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

#### **Gráficos 3D**

También existe la posibilidad de visualizar los datos en tres dimensiones.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# Datos para linea 3d
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Datos para dispersion 3d
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

Podemos hacer lo mismo para contornos:

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

Si queremos cambiar el ángulo de visualización, podemos hacerlo con `view_init`, ajustando el grado de elevación y el de rotación:

```python
ax.view_init(60, 35)
fig
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go

# Crear la figura 3D interactiva
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])

# Configurar el diseño de la figura
fig.update_layout(
    title='Gráfico de Superficie 3D Interactivo',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z = f(X, Y)'
    )
)

# Mostrar la figura interactiva
fig.show()
```

### 2. Limpieza y transformación de datos

A veces, los datos no están en el formato más conveniente para el análisis. Por ejemplo, las columnas de tiempo pueden estar como texto y necesitamos convertirlas a un formato numérico:


```python
import datetime

def convert_time(s):
    h, m, s = map(int, s.split(':'))
    return datetime.timedelta(hours=h, minutes=m, seconds=s)

data = pd.read_csv('marathon-data/marathon-data.csv',
                   converters={'split':convert_time, 'final':convert_time})
data.head()
```

```python
data.dtypes
```

Creamos nuevas columnas con los tiempos en segundos para facilitar el análisis:

```python
data['split_sec'] = data['split'].view(int) / 1E9
data['final_sec'] = data['final'].view(int) / 1E9
data.head()
```

### 4. Visualización exploratoria

#### Distribución de tiempos

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(data['final_sec']/3600, bins=30, kde=True)
plt.xlabel('Tiempo final (horas)')
plt.title('Distribución de tiempos finales en el maratón')
plt.show()
```

#### Relación entre split y tiempo final

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='split_sec', y='final_sec', data=data, alpha=0.3)
plt.xlabel('Tiempo en el split (segundos)')
plt.ylabel('Tiempo final (segundos)')
plt.title('Relación entre split y tiempo final')
plt.show()
```

#### Diferencias por género

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='gender', y='final_sec', data=data)
plt.xlabel('Género')
plt.ylabel('Tiempo final (segundos)')
plt.title('Distribución de tiempos finales por género')
plt.show()
```

#### Análisis de correlaciones

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matriz de correlación")
plt.show()
```

#### Comparación de tiempos por grupos de edad

Agrupa por rangos de edad y compara los tiempos finales:

```python
data['age_group'] = pd.cut(data['age'], bins=[10,20,30,40,50,60,70,80], right=False)
plt.figure(figsize=(12,6))
sns.boxplot(x='age_group', y='final_sec', data=data)
plt.xlabel('Grupo de edad')
plt.ylabel('Tiempo final (segundos)')
plt.title('Tiempos finales por grupo de edad')
plt.show()
```
